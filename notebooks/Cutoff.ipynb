```python
import pandas as pd
import numpy as np
import time
from datetime import datetime, timedelta
```

```python
import os
```

```python
inputdir = ["2015","2016","2017","2018","2019","2020","2021"]
```

```python
li = []
for item in inputdir:
    df = pd.read_csv("../Data/Archive/PriceResult/"+item+"_price.csv",parse_dates=["datetime"],index_col="datetime")
    li.append(df)
data = pd.concat(li, axis=0)
```

```python
data
```

```python
data.index.max()
```

```python
cutoff_df = data.groupby(pd.Grouper(freq='M')).apply(lambda x: (x.index.max() + timedelta(days=-2)).replace(hour=13, minute=29, second=0)).to_frame("cutoff_date")
```

```python
cutoff_date = cutoff_df["cutoff_date"].astype(str).tolist()
```

```python
cutoff_pair = list(zip(cutoff_date, cutoff_date[1:]))
cutoff_pair
```

```python
cut_monty = cutoff_df.index.strftime('%Y%m').tolist()
```

```python
cut_monty.append('202106')
```
